```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Data.csv', encoding='ISO-8859-1')
```

```python
data.head() # we have top twenty headline news and with this we need to predict whether the stock will go up or down.
```

```python
# splitting into train and test
train = data[data['Date']< '20150101']
test = data[data['Date']> '20141231']
```

```python
train.head()
```

```python
test.head()
```

```python
train.columns
```

```python
# preprocessing the dataset
df = train.iloc[:,2:27]
df.replace("^[a-zA-Z]", " ", regex=True, inplace=True)
```

```python
# renaming coumn name for ease of access
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
df.columns = new_index
df.head(5)
```

```python
# converting headlines to lower case
for index in new_index:
  df[index] = df[index].str.lower()
df.head(1)
```

```python
df.index
```

```python
df.columns
```

```python
" ".join(str(x) for x in df.iloc[1,0:25]) # this is how we can join every headline of the first row and make it in a paragraph.
```

```python
# here we did the similiar thing of joining but for all rows
headlines = []
for row in range(0, len(df.index)):
  headlines.append(" ".join(str(x) for x in df.iloc[row,0:25]))
```

```python
headlines
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
# implementing bag of words
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
# implementing random forest classifier
random_classifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
random_classifier.fit(traindataset, train['Label'])
```

```python
# predict for the test dataset
test_transform = []
for row in range(0, len(test.index)):
  test_transform.append(" ".join(str(x) for x in test.iloc[row, 2:27]))
test_dataset = countvector.transform(test_transform) # converting to bag of words
predictions = random_classifier.predict(test_dataset)
```

```python
predictions
```

```python
# let's check the accuracy of the predictions
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

matrix = confusion_matrix(test['Label'], predictions)
print(matrix)
print('\n\n')
score = accuracy_score(test['Label'], predictions)
print(score)
print('\n\n')
report = classification_report(test['Label'], predictions)
print(report)
```
